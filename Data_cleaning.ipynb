## Cleaning data

```python
df = pd.read_csv(raw, index_col=0)
```

```python
df.head()
```

```python
df.shape
```

```python
# yields ValueError: ValueError: could not convert string to float: 'STD'
#df['Depth (ft)'].astype('float64')
```

```python
df['Depth (ft)'].str.contains('STD', na=False)
```

```python
df[df['Depth (ft)'].str.contains('STD', na=False)]
```

```python
df[df['Depth (ft)'].str.contains('-', na=False)]
```

```python
df.loc[df['Depth (ft)'].str.contains('-', na=False), 'Depth (ft)'] = np.nan
df.loc[df['Depth (ft)'].str.contains('STD', na=False), 'Depth (ft)'] = np.nan
```

```python
df['Depth (ft)'] = df['Depth (ft)'].astype('float64')
```

```python
df['Depth (ft)'].dtypes
```

```python
#df['Sample ID'].astype('float64')
```

```python
df.loc[df['Sample ID'].str.contains('-', na=False), 'Sample ID'] = np.nan
```

```python
df['Sample ID'] = df['Sample ID'].astype('float64')
```

```python
# yields ValueError: could not convert string to float: '836.0 0.0000E+00'
#df.iloc[:, 7].astype('float64')
```

```python
df.columns[7]
```

```python
df.loc[df['AMU1'].str.contains('E+', regex=False, na=False) == True]
```

```python
#df[df['Well ID'] == '24A-8'][-50:]
```

```python
df.loc[(df['AMU1'].str.contains('E+', regex=False, na=False) == True) &
       (df['Well ID'] == '24A-8'), 'AMU1'] = np.nan
```

```python
df.loc[df['Well ID'] == '24A-8', 'AMU1'] = np.nan
```

```python
df.loc[df['Well ID'] == '24A-8', 'AMU1'].astype('float64')
```

```python
df.to_csv('./data/cleaned_types.csv', index=False)
```
